Numpy Indexing and Selection

```python
arr = np.arange(0,11) #11 elements from 0 to 10
arr
```

```python
arr[0]
```

```python
arr[1:4]
```

```python
arr[:6]
```

```python
arr[5:]
```

```python
arr[0:5] = 100 #Broadcast the value 100 to 1st 5 elements
arr
```

```python
arr = np.arange(0,11)
arr
```

```python
slice_of_arr = arr[0:6]
slice_of_arr
```

```python
arr1 = slice_of_arr[:] # They are the same objects
arr1[:] = 10
arr1
slice_of_arr
arr1 == slice_of_arr
```

```python
arr1 = slice_of_arr.copy() #Copy object not same
arr1[:] = 4
arr1
slice_of_arr
arr1 == slice_of_arr
```

#Indexing 2D Array/Matrix

```python
arr_2d = np.array([[5,10,15],[20,25,30],[35,40,45]])
arr_2d
```

```python
#Double Bracket Format
arr_2d[0][0] #Returns digit in first row and first column 
```

```python
arr_2d[0] #Retuns the entire first row. [R][C]
```

```python
arr_2d[1][1]
arr_2d[2][1]
```

```python
#Single Bracket Notation [R,C]
arr_2d[1,2]
```

```python
#Subsetting
arr_2d[1:,1:]
arr_2d[:2,:1]
```

```python
arr_2d[:2]
```

```python
arr_2d[:,:1]
```
