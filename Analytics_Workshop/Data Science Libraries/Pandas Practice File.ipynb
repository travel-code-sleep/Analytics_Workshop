# Manage Missing Data

```python
d = {'A':[1,2,np.nan],'B':[5,np.nan,np.nan],'C':[1,2,3]}
d
```

```python
df = pd.DataFrame(d)
df #Nans are missing values
```

```python
df.dropna() # drops rows with Nan
```

```python
df.dropna(axis=1) #drops columns with Nan
```

```python
df #remember no inplace change
```

```python
#specify threshold 
df.dropna(thresh=2) #drops any row with two or more non NaN values
```

```python
#Replacing missing values
df.fillna(value=df.mean()) #feeding mean of dataframe
```

```python
df['A'].fillna(value=df['A'].mean()) #feeding feature wise mean 
```

# Meging, Joining and Concatneting 


```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3],axis=1) #concat across axis. will create NaN values to fill empty positions. Observe the pattern
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})    
```

```python
left, right
```

```python
#Merging dfs
pd.merge(left,right,how="inner",on="key") #similar to sql joins. "how" defines type of join, "on" defines constraint columns
```

```python
left1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right1 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
pd.merge(left1,right1,how='inner',on=['key1','key2']) #merge using 2 features
```

```python
pd.merge(left1,right1,how='outer',on=['key1','key2'])
```

```python
left2 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right2 = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
#joining
left2.join(right2) #uses the index feature instead of keys to merge two data frames
```

```python
left2.join(right2,how='outer') 
```
