```python
x = np.linspace(0,5,11)
y = x**2
```

```python
y
```

```python
#Functional plotting
plt.plot(x,y,'r-')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('X vs. Y')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#OO Plotting
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('X vs. Y')
```

```python
#Multiple Figures one one canvas
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.4,0.3,0.4])
axes1.plot(x,y)
axes1.set_title('X vs. Y')
axes2.plot(y,x)
axes2.set_title('Y vs. X')
```

```python
#Setting Graph Legend 
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X vs X^2')
ax.plot(x,2**x,label='X vs. 2^X')
ax.legend()
#ax.legend(loc=(.5,.5))
```

```python
#Plot Customization
fig = plt.figure().add_axes([0,0,1,1])
#fig.plot(x,y,color='#FF8C00') #RGB Hex Code
#fig.plot(x,y,color='purple', linewidth=3) #edit lindewidth
#fig.plot(x,y,color='purple', linewidth=3, alpha=0.5) #transparency
#fig.plot(x,y,color='purple', lw=3, ls='steps') #Line Style
fig.plot(x,y,color='purple', lw=1, ls='--',marker='*',markersize=20,markerfacecolor='orange',markeredgewidth=1.5,
         markeredgecolor='black')
```

```python
#Set X & Y Limit
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,ls='--')
ax.set_xlim([0,2])
ax.set_ylim([0,10])
```
